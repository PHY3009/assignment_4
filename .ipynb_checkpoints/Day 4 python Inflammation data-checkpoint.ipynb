```python
#saving the inflammation[ * ].csv files by naming them as a new variable

file_names = glob.glob('inflammation*.csv')

for files in file_names:
    
    print(files)
    
    data=numpy.loadtxt(fname=files, delimiter= ',')

fig = matplotlib.pyplot.figure(figsize=(10.0, 3.0)) #10 is width, 3 is height

subplot_1 = fig.add_subplot(1, 3, 1)
subplot_2 = fig.add_subplot(1, 3, 2)
subplot_3 = fig.add_subplot(1, 3, 3)

subplot_1.set_ylabel('average inflammation')
subplot_1.plot(data.mean(axis=0))

subplot_2.set_ylabel('maximum inflammation')
subplot_2.plot(data.max(axis=0))

subplot_3.set_ylabel('minimum inflammation')
subplot_3.plot(data.min(axis=0))

fig.tight_layout() #gives figures breathing room

matplotlib.pyplot.show(fig)
```

```python
def fahr_to_kelvin(f_temp):
    k_temp = (f_temp -32)*(5/9)+273.15
    return k_temp
```

```python
boiling_temp = farh_to_kelvin(32)
print(boiling_temp)
```

```python
def kelvin_to_celcius(k_temp):
    c_temp = k_temp - 273.15
    return c_temp

absolute_zero = kelvin_to_celcius(0) #when k_temp=0
print('absolute zero in Celcius is', absolute_zero)
```

```python
def fahr_to_celcius(f_temp):
    temp_k = fahr_to_kelvin(f_temp)
    temp_c = kelvin_to_celcius(temp_k)
    return temp_c
```

```python
freezing_temp = fahr_to_celcius(32) #when Farenheit is 32, celcius should be 'freezing_temp'
print(freezing_temp)
```

```python
#If I hit 34 and miss 21 targets, what % of targets were hit?

#provide variables
targets_hit = 34
targets_missed = 21

#create function to calculate proportion from absolute number
def proportion(success, failure):
    percent_hit = success/ (success+failure)
    return percent_hit #what you want to get out

print(proportion(targets_hit, targets_missed))
```
