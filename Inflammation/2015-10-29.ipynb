```python
#lets make a function of changing F to K (temp) 
def fahr_to_kelvin(ftemp):
#def tells python we are defining a function, then we need the name of function, then possible input. Can have
#multiple is we seperate them with commas.
    k_temp = (ftemp - 32)*(5/9) + 273.15
    return k_temp
    
```

```python
boiling_temp = fahr_to_kelvin(32) 
print (boiling_temp)
```

```python
def kelvin_to_celcius (k_temp):
    c_temp = k_temp - 273.15
    return c_temp
```

```python
absolute_zero = kelvin_to_celcius(0)
print (absolute_zero)
```

```python
def fahr_to_celsius (f_temp):
#we can use the functions from earlier to use make this simpler
    temp_k = fahr_to_kelvin(f_temp)
    temp_c = kelvin_to_celcius(temp_k)
    return temp_c
```

```python
#try to print the freezing temperature of water

freezing_temp = fahr_to_celsius(32)
print (freezing_temp)
```

```python
def proportion (successes, failures):
    total = successes + failures
    my_proportion = successes / total
    return my_proportion
```

```python
targets_hit = 34
targets_missed = 21

print (proportion(targets_hit,targets_missed))
```

```python
#function to calculate to convert days to years
def days_to_years (days):
    '''Function to convert days to years
    Input: days
    Output: years'''
    #explain the math in this comment
    years = days/365
    return years
```
